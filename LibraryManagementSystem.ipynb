```python
# =========================================================
# LIBRARY MANAGEMENT SYSTEM
# =========================================================

# Import CSV module for file handling
import csv

# Import OS module to check files
import os

# File names
BOOK_FILE = "books.csv"
ISSUE_FILE = "issued_books.csv"
```

```python
# =========================================================
# FUNCTION TO CREATE FILES
# =========================================================

def create_files():

    # Check if books.csv exists
    if not os.path.exists(BOOK_FILE):

        # Create books.csv file
        with open(BOOK_FILE, "w", newline="") as file:

            writer = csv.writer(file)

            # Add column headings
            writer.writerow([
                "Book ID",
                "Book Name",
                "Author",
                "Quantity"
            ])

    # Check if issued_books.csv exists
    if not os.path.exists(ISSUE_FILE):

        # Create issued_books.csv file
        with open(ISSUE_FILE, "w", newline="") as file:

            writer = csv.writer(file)

            # Add column headings
            writer.writerow([
                "Student Name",
                "Book ID",
                "Book Name"
            ])
```

```python
# =========================================================
# FUNCTION TO ADD BOOK
# =========================================================

def add_book():

    try:
        # Take input from user
        book_id = input("Enter Book ID: ")
        book_name = input("Enter Book Name: ")
        author = input("Enter Author Name: ")

        # Convert quantity into integer
        quantity = int(input("Enter Quantity: "))

        # Open file in append mode
        with open(BOOK_FILE, "a", newline="") as file:

            writer = csv.writer(file)

            # Add new book record
            writer.writerow([
                book_id,
                book_name,
                author,
                quantity
            ])

        print("Book Added Successfully!\n")

    # Handle invalid quantity
    except ValueError:
        print("Invalid Quantity!\n")
```

```python
# =========================================================
# FUNCTION TO DISPLAY BOOKS
# =========================================================

def display_books():

    # Open books.csv file
    with open(BOOK_FILE, "r") as file:

        reader = csv.reader(file)

        # Skip header row
        next(reader)

        print("\n========== BOOK RECORDS ==========")

        # Read all records
        for row in reader:

            print("Book ID :", row[0])
            print("Book Name :", row[1])
            print("Author :", row[2])
            print("Quantity :", row[3])

            print("--------------------------------")
```

```python
# =========================================================
# FUNCTION TO ISSUE BOOK
# =========================================================

def issue_book():

    # Input student name
    student_name = input("Enter Student Name: ")

    # Input book ID
    book_id = input("Enter Book ID to Issue: ")

    books = []

    found = False

    # Open books file
    with open(BOOK_FILE, "r") as file:

        reader = csv.reader(file)

        # Store header
        header = next(reader)

        # Read all records
        for row in reader:

            # Match Book ID
            if row[0] == book_id:

                found = True

                # Check quantity
                if int(row[3]) > 0:

                    # Reduce quantity
                    row[3] = str(int(row[3]) - 1)

                    # Store issue record
                    with open(ISSUE_FILE, "a", newline="") as issue_file:

                        writer = csv.writer(issue_file)

                        writer.writerow([
                            student_name,
                            row[0],
                            row[1]
                        ])

                    print("Book Issued Successfully!\n")

                else:
                    print("Book Not Available!\n")

            # Store updated rows
            books.append(row)

    # If Book ID not found
    if not found:
        print("Book ID Not Found!\n")

    # Rewrite updated data
    with open(BOOK_FILE, "w", newline="") as file:

        writer = csv.writer(file)

        writer.writerow(header)

        writer.writerows(books)
```

```python
# =========================================================
# FUNCTION TO RETURN BOOK
# =========================================================

def return_book():

    # Input Book ID
    book_id = input("Enter Book ID to Return: ")

    books = []

    # Open books file
    with open(BOOK_FILE, "r") as file:

        reader = csv.reader(file)

        # Store header row
        header = next(reader)

        # Read records
        for row in reader:

            # Match Book ID
            if row[0] == book_id:

                # Increase quantity
                row[3] = str(int(row[3]) + 1)

                print("Book Returned Successfully!\n")

            books.append(row)

    # Rewrite updated records
    with open(BOOK_FILE, "w", newline="") as file:

        writer = csv.writer(file)

        writer.writerow(header)

        writer.writerows(books)
```

```python
# =========================================================
# FUNCTION TO SEARCH BOOK
# =========================================================

def search_book():

    # Input search value
    search = input(
        "Enter Book Name or Author Name: "
    ).lower()

    found = False

    # Open books file
    with open(BOOK_FILE, "r") as file:

        reader = csv.reader(file)

        # Skip heading row
        next(reader)

        # Search records
        for row in reader:

            # Match book name or author
            if search in row[1].lower() or \
               search in row[2].lower():

                print("\nBook Found!")

                print("Book ID :", row[0])
                print("Book Name :", row[1])
                print("Author :", row[2])
                print("Quantity :", row[3])

                found = True

    # If no book found
    if not found:
        print("No Book Found!\n")
```

```python
# =========================================================
# FUNCTION TO DISPLAY REPORTS
# =========================================================

def reports():

    available_books = 0

    issued_books = 0

    # Open books file
    with open(BOOK_FILE, "r") as file:

        reader = csv.reader(file)

        # Skip header
        next(reader)

        # Count available books
        for row in reader:

            available_books += int(row[3])

    # Open issued books file
    with open(ISSUE_FILE, "r") as file:

        reader = csv.reader(file)

        # Skip header
        next(reader)

        # Count issued books
        for row in reader:

            issued_books += 1

    # Display report
    print("\n========== LIBRARY REPORT ==========")

    print("Available Books :", available_books)

    print("Issued Books :", issued_books)
```

```python
# =========================================================
# MAIN MENU FUNCTION
# =========================================================

def main():

    # Create files
    create_files()

    while True:

        # Display menu
        print("\n========== LIBRARY MANAGEMENT SYSTEM ==========")

        print("1. Add Book")
        print("2. Display Books")
        print("3. Issue Book")
        print("4. Return Book")
        print("5. Search Book")
        print("6. Reports")
        print("7. Exit")

        # Take user choice
        choice = input("Enter Your Choice: ")

        # Call functions
        if choice == "1":
            add_book()

        elif choice == "2":
            display_books()

        elif choice == "3":
            issue_book()

        elif choice == "4":
            return_book()

        elif choice == "5":
            search_book()

        elif choice == "6":
            reports()

        elif choice == "7":

            print("Thank You!")

            break

        else:
            print("Invalid Choice!\n")
```

```python
# =========================================================
# PROGRAM STARTS HERE
# =========================================================

main()
```
